```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as date
import random
from sklearn.cluster import KMeans
```

```python
genders = []
information = pd.read_csv("PSIM/500.csv")
# for val in information['Gender']:
#     if (val == 'Male'):
#         genders.append(1)
#     else:
#         genders.append(0)
# information['Gender'] = genders
information = information.drop(['Gender'], axis=1)
information = information.drop(['Index'], axis=1)
information.head()
```

```python
f1 = information['Weight'].values
f2 = information['Height'].values
x = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, s=7)
```

```python
kmeans5 = KMeans(n_clusters=5)
y_kmeans5 = kmeans5.fit_predict(x)
print(y_kmeans5)
```

```python
kmeans5.cluster_centers_
```

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(x)
    kmeans.fit(x)
    Error.append(kmeans.inertia_)
import matplotlib.pyplot as plt
plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```
